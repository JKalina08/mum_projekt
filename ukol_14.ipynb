### Histogram, jeho aplikace a výpovědní hodnota.

#### __Úvod__  
Histogram je grafická reprezentace dat, která zobrazuje rozdělení hodnot v určité skupině dat. Vyznačuje se tím, že hodnoty jsou rozděleny do předem určených binů (intervalů) a na ose y je zobrazena frekvence, tedy počet hodnot v daném binu. Histogram se často používá k vyjádření distribuce hodnot v nějakém datasetu a k porovnání s jinými distribucemi. Může být také užitečný při hledání anomálií nebo odchylek v datech.

#### __Motivace a hypotézy__  
Histogram grayscale obrázku je grafické zobrazení rozložení jasových hodnot v obrázku. Může být užitečné pro úpravu kontrastu a jasu obrázku a pro detekci expozičních chyb.

Funkce __Levels__ se používá k nastavení minimálního a maximálního jasu v obrázku a k vytvoření křivky, která přepočítává všechny jasové hodnoty v obrázku na nové hodnoty podle zvolené křivky. Tato funkce může být užitečná pro zlepšení kontrastu a úpravu jasu obrázku. Po aplikaci funkce Levels se histogram obrázku změní tak, že se rozložení hodnot přesune k vybraným minimálnímu a maximálnímu jasu.

Funkce __Curves__ se podobá funkci __Levels__, ale umožňuje více přesné nastavení křivky pomocí více bodů na křivce. Tato funkce také umožňuje nastavit různé křivky pro jednotlivé barvy nebo pro různé oblasti jasu v obrázku. Po aplikaci funkce __Curves__ se histogram obrázku změní podle nově nastavené křivky.

Funkce __Brightness&Contrast__ se používá k úpravě jasu a kontrastu obrázku pomocí jednoduchého ovládání. Po aplikaci této funkce se histogram obrázku změní tak, že se rozložení hodnot přesune podle nastavení jasu a kontrastu.

Funkce __Shadows/Highlights__ se používá k úpravě stínu a světla v obrázku. Umožňuje nastavit, jak moc se mají stíny a světla zesvětlit nebo ztmavit. Po aplikaci této funkce se histogram obrázku může změnit v závislosti na tom, jak byly stíny a světla upravena. 

Funkce __Exposure__ se používá k úpravě celkové expozice grayscale obrázku. Umožňuje nastavit, jak moc se má obrázek zesvětlit nebo ztmavit. Po aplikaci této funkce se histogram obrázku změní tak, že se rozložení hodnot přesune podle nastavení expozice.

Při přeexponování grayscale obrázku dochází k příliš vysokému zastoupení jasných hodnot v obrázku, což může vést k tomu, že se detaily ve světlech ztratí. Histogram přeexponovaného obrázku bude mít výrazný posun ke světlejším hodnotám. Při podexponování obrázku dochází naopak k příliš vysokému zastoupení tmavých hodnot, což může vést k tomu, že se detaily ve stínech ztratí. Histogram podexponovaného obrázku bude mít výrazný posun ke tmavším hodnotám.

Expoziční vady lze částečně napravit pomocí funkcí __Levels__, __Curves__, __Brightness&Contrast__ a __Exposure__. Je důležité si uvědomit, že úpravy expozice mohou vést k dalším problémům, jako je například ztráta kontrastu nebo nadměrné žloutnutí nebo modření obrázku (v případě barevných obrázků). Je třeba tedy úpravy expozice provádět opatrně a zvážit všechny možné důsledky.

Histogram, jeho aplikace a výpovědní hodnota. U grayscale obrázků demonstrujte, jak se změní obrázek a jeho histogram po aplikaci funkcí Levels, Curves, Brightness&Contrast, Shadows/Highlight a Exposure. Použijte obrázek s plynulým přechodem z černé do bílé a kvalitní čb foto. Na čb fotu demonstrujte, jak se na histogramu projeví přeexponování a podexponování. 


```python
from matplotlib import pyplot as plt
from skimage import io
import numpy as np
%matplotlib inline

BINS = 256
# Loading sample image from disc
# Original image
dog = io.imread('./imgs/dog.jpg')
dog.astype(np.uint8)
# Load image of dog after applying "Levels" in GIMP
# Levels
dog_levels = io.imread('./imgs/dog_levels.jpg')
dog_levels.astype(np.uint8)
# Curves
dog_curves = io.imread('./imgs/dog_curves.jpg')
dog_curves2 = io.imread('./imgs/dog_curves_hist.jpg')
dog_curves.astype(np.uint8)
# Brightness & Contrast
dog_brcon = io.imread('./imgs/dog_brcon.jpg')
dog_brcon2 = io.imread('./imgs/dog_brcon_hist.jpg')
dog_brcon.astype(np.uint8)
# Shadows/Highlights
dog_shahig = io.imread('./imgs/dog_shahig.jpg')
dog_shahig2 = io.imread('./imgs/dog_shahig_hist.jpg')
dog_shahig.astype(np.uint8)
# Exposure
dog_expo = io.imread('./imgs/dog_expo.jpg')
dog_expo2 = io.imread('./imgs/dog_expo_hist.jpg')
dog_expo.astype(np.uint8)

# Plotting the images
# Create the figure and subplots
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(13, 13),
                         gridspec_kw={'height_ratios': [2,1,2,1]})
# Original
axes[0,0].imshow(dog, cmap=plt.cm.gray)
axes[0,0].set_title("Obr. 1 - Originál")
axes[1,0].hist(dog.ravel(), BINS)
axes[1,0].set_title("Obr. 1.1 - Orig. histogram")
# Levels
axes[0,1].imshow(dog_levels, cmap=plt.cm.gray)
axes[0,1].set_title("Obr. 2 - Levels")
axes[1,1].hist(dog_levels.ravel(), BINS, color='red')
axes[1,1].set_title("Obr. 2.1 - Levels histogram")
# Curves
axes[0,2].imshow(dog_curves, cmap=plt.cm.gray)
axes[0,2].set_title("Obr. 3 - Curves")
axes[1,2].hist(dog_curves.ravel(), BINS, color='green')
axes[1,2].set_title("Obr. 3.1 - Curves histogram")
#  Brightness & Contrast
axes[2,0].imshow(dog_brcon, cmap=plt.cm.gray)
axes[2,0].set_title("Obr. 4 - Bright.&Contr.")
axes[3,0].hist(dog_brcon.ravel(), BINS, color='cyan')
axes[3,0].set_title("Obr. 4.1 - Bright.&Contr. histogram")
# Shadows/Highlights
axes[2,1].imshow(dog_shahig, cmap=plt.cm.gray)
axes[2,1].set_title("Obr. 5 - Shadows/Highlights")
axes[3,1].hist(dog_shahig.ravel(), BINS, color='blue')
axes[3,1].set_title("Obr. 5.1 - Shadows/Highlights histogram")
# Exposure
axes[2,2].imshow(dog_expo, cmap=plt.cm.gray)
axes[2,2].set_title("Obr. 6 - Exposure")
axes[3,2].hist(dog_expo.ravel(), BINS, color='m')
axes[3,2].set_title("Obr. 6.1 - Exposure histogram")

for n_row, row in enumerate(axes):
    for n_col, col in enumerate(row):
        if n_row in (0, 2):
            col.set_xticks([])
            col.set_yticks([])
        else:
            col.set_ylabel("Počet pixelů")
            col.set_xlabel("Intenzita pixelu")

plt.tight_layout()
plt.show()
```
